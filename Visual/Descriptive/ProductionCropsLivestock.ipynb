```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import StructType,StructField, StringType, IntegerType
import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import collections
```

```python
COUNTRY = 'Viet Nam'
```

```python
# spark = SparkSession.builder \
#                     .master("spark://127.0.0.1:7077") \
#                     .appName('ProductionCropsLivestock') \
#                     .getOrCreate()
spark = SparkSession.builder \
                    .master("local[1]") \
                    .appName('ProductionCropsLivestock') \
                    .getOrCreate()
```

```python
pcl_df = spark.read\
    .parquet("hdfs://127.0.0.1:9000/FAOSTAT_prj/DataWarehouse/Production_Crops_Livestock")
```

```python
ElementProd_df = spark.read\
    .parquet("hdfs://127.0.0.1:9000/FAOSTAT_prj/DataWarehouse/Element_Prod")
```

```python
ItemProd_df = spark.read\
    .parquet("hdfs://127.0.0.1:9000/FAOSTAT_prj/DataWarehouse/Item_Prod")
```

```python
country_df = spark.read\
    .parquet("hdfs://127.0.0.1:9000/FAOSTAT_prj/DataWarehouse/Country")
```

```python
final_df = pcl_df.join(country_df, pcl_df.country_id == country_df.id, "inner") \
                .filter(country_df.country == COUNTRY) \
                .join(ElementProd_df, pcl_df.element_prod_id ==  ElementProd_df.id, "inner") \
                .join(ItemProd_df, pcl_df.item_prod_id ==  ItemProd_df.id, "inner") \
                .drop('country_id', 'item_prod_id', 'element_prod_id', 'flag', 'transformed_date', 'id', 'sub_region_code', 'region_code', 'id', 'sub_region_name', 'region_name', 'country') \
                .toPandas()
```

```python
display(final_df)
```

```python
final_df.year = final_df.year.astype(int)
final_df.value = final_df.value.astype(float)
print(final_df.info())
```

```python
final_CROPS_PRIMARY_df = final_df[final_df.item_category == 'CROPS_PRIMARY']

final_CROPS_PRIMARY_ha_df = final_CROPS_PRIMARY_df[final_CROPS_PRIMARY_df.unit == 'ha']
final_CROPS_PRIMARY_tonnes_df = final_CROPS_PRIMARY_df[final_CROPS_PRIMARY_df.unit == 'tonnes']
```

```python
def plot_crops_livestock(df, label):
    df.loc[:, 'rank_per_year'] = df.groupby('year')['value'] \
                                                                                .rank(method='min', ascending=False)

    pcl_list = df[df.rank_per_year == 1.0].sort_values('year').item.tolist()
    pcl_2nd_list = df[df.rank_per_year == 2.0].sort_values('year').item.tolist()

    filtered_ha_items = []

    for item in pcl_list:
        if 'Other' not in item:
            filtered_ha_items.append(item)

    rank_1st_first_year = filtered_ha_items[0]
    rank_1st_last_year = filtered_ha_items[-1]

    item_counts = collections.Counter(filtered_ha_items)
    three_most_frequent_items = item_counts.most_common(3)

    visualization_item = [rank_1st_first_year, rank_1st_last_year]
    visualization_item = visualization_item + [item[0] for item in three_most_frequent_items]

    filtered_ha_items = []

    for item in pcl_2nd_list:
        if 'Other' not in item:
            filtered_ha_items.append(item)

    rank_2nd_first_year = filtered_ha_items[0]
    rank_2nd_last_year = filtered_ha_items[-1]

    item_counts = collections.Counter(filtered_ha_items)
    three_most_frequent_items = item_counts.most_common(3)

    visualization_item = visualization_item + [rank_2nd_first_year, rank_2nd_last_year]
    visualization_item = visualization_item + [item[0] for item in three_most_frequent_items]

    unique_set = set(visualization_item)
    visualization_item = list(unique_set)

    data = df.loc[df['item'].isin(visualization_item), ['item', 'year', 'value']].sort_values('year')

    year_list = np.array(pd.unique(data.year.astype('int').tolist()))

    for i in range(min(len(visualization_item), 5)):
        plt.plot(year_list, np.array(data[data['item'] == visualization_item[i]].value.fillna(0).round(2).tolist()), label = visualization_item[i])

    plt.xlabel("Year")
    plt.ylabel("ha")
    plt.legend()
    plt.title(label)
```

```python
plot_crops_livestock(final_CROPS_PRIMARY_ha_df, 'The CROPS PRIMARY which have the largest area')
```

```python
plot_crops_livestock(final_CROPS_PRIMARY_tonnes_df, 'The CROPS PRIMARY which have the largest production')
```
